```python
import watermark
from watermark import config
```

```python
from torch.utils.data import DataLoader
```

```python
train_images_paths = list(config.paths.RAW.glob("train/*"))
test_images_paths = list(config.paths.RAW.glob("train/*"))
logo_paths = list(config.paths.LOGOS.glob("*"))
```

```python
train_dataset = watermark.WatermarkDataset(
    watermarks=[
        watermark.LogoWatermark(
            logo_path=logo_path, percentage_size=15, intensity=125, randomize=True
        )
        for logo_path in logo_paths
    ],
    image_paths=train_images_paths,
)
test_dataset = watermark.WatermarkDataset(
    watermarks=[
        watermark.LogoWatermark(
            logo_path=logo_path, percentage_size=15, intensity=125, randomize=True
        )
        for logo_path in logo_paths
    ],
    image_paths=test_images_paths,
)
```

```python
train_dataloader = DataLoader(dataset=train_dataset, batch_size=16, shuffle=True)
test_dataloader = DataLoader(dataset=test_dataset, batch_size=16, shuffle=True)
```
